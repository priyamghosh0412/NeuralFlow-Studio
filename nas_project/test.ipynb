```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("db.sqlite3")
cursor = conn.cursor()
```

```python
table_names = [
    "django_migrations",
    "sqlite_sequence",
    "auth_group_permissions",
    "auth_user_groups",
    "auth_user_user_permissions",
    "django_admin_log",
    "django_content_type",
    "auth_permission",
    "auth_group",
    "auth_user",
    "django_session",
]

for table_name in table_names:
    print("table_name: ",table_name)
    df = pd.read_sql_query(f"SELECT * FROM {table_name} LIMIT 10", conn)
    print(df)
    print("\n")
```
